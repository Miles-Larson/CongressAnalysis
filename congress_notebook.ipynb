```python
import secret_api
import requests
api_key = secret_api.congress_gov_api


def call_bills_api (api_key):
    file_type = 'json'
    limit_cap = str(250)
    offset_iter = 0

    url = "https://api.congress.gov/v3/bill?format=" + file_type +" &offset=" + str(offset_iter) + "limit="+ limit_cap + "&sort=updateDate%2Basc&api_key=" + api_key
    headers = {"accept": "application/json"}

    response = requests.get(url,headers=headers)
    return response.text


    response.raise_for_status()




```
